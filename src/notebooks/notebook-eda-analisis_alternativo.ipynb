```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

## Análisis bivariante ##
Análisis alternativo de **stroke** con las variables *ever_married, work_type y Residence_type*, para valorar si hay una relación directa:

```python
# Cargamos el df para visualizarlo:
df_stroke_prediction = pd.read_csv("../data/healthcare-dataset-stroke-data.csv")
df_stroke_prediction.head(10)
df_stroke = df_stroke_prediction.copy()
```

```python
sns.catplot(x="stroke",
            hue = "ever_married",
            kind="count",
            edgecolor=".6",
            orient = "V",
            data=df_stroke);
# fig.savefig("../img/bivariantestroke_ever_married_barras.png")
```

- Variable *stroke* con *ever_married*:

La mayoría de las personas que han tenido ictus, han estado casadas. 

```python
sns.catplot(x="stroke",
            hue = "work_type",
            kind="count",
            edgecolor=".6",
            orient = "V",
            data=df_stroke);
# fig.savefig("../img/bivariante_stroke_work_type_barras.png")
```

- Variable *stroke* con *work_type*:

La mayoría de las personas que han tenido ictus, han trabajado en el sector privado. 

```python
sns.catplot(x="stroke",
            hue = "Residence_type",
            kind="count",
            edgecolor=".6",
            orient = "V",
            data=df_stroke);
# fig.savefig("../img/bivariante_stroke_residence_type_barras.png")
```

- Variable *stroke* con *Residence_type*:

La mayoría de las personas que han tenido ictus, han vivido en zonas urbanas. 

### Conclusión ###

El perfil de persona que ha tenido un ictus han estado casados, trabajado en el sector privado, y residido en zonas urbanas. Asimismo, el porcentaje es muy bajo y no encontramos una relación directa con tener ictus.
